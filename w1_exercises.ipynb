```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# 1. Use .loc to get the name and age of the passenger with index 12
print(df.loc[12, ["Age", "Name"]])

# 2. Use .iloc to get the first 5 rows and first 3 columns
print(df.iloc[0:6, 0:3])

# 3. Use .loc to get all passengers in 1st class who are female
print(df.loc[(df["Pclass"] == 1) & (df["Sex"] == "female")])

# 4. Use .iloc to get the age of the 10th passenger (by position)
print(df.iloc[9, 5])

# 5. Use .loc to get the 'Fare' and 'Survived' values for passengers with index 100 to 105
print(df.loc[100:105, ["Fare", "Survived"]])
```

Section 2 : Clean & Tidy Data

Exercise Set 1 : Dropping Data
Exercises 1-4

```python
import pandas as pd

data = {
"Name": ["Alice", "Bob", "Charlie", "David"],
"Age": [25, 30, 35, 40],
"City": ["NYC", "LA", "Chicago", "Miami"],
"Email": ["a@example.com", "b@example.com", "c@example.com", "d@example.com"],
}

df = pd.DataFrame(data)

# 1. Drop the Email column and store the result in a new variable.
df_no_email = df.drop(['Email'], axis=1)
print(df_no_email)

# 2. Drop the row with index 2 in place.
df = df.drop(2, axis=0)
print(df)

# 3. Drop the City and Email columns, and assign the result back to df.
df = pd.DataFrame(data)
df = df.drop(['City', 'Email'], axis=1)
print(df)

# 4. Drop the first and last row, and store the result in df_trimmed.
df = pd.DataFrame(data)
df_trimmed = df.drop([df.index[0], df.index[-1]], axis=0)
print(df_trimmed)
```

Exercise Set 2 : Missing Data
Exercises 1-9

```python
import pandas as pd
import numpy as np

data = {
"Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
"Age": [25, np.nan, 35, 40, None],
"City": ["NYC", "LA", None, "Miami", "Boston"],
}

df = pd.DataFrame(data)

# 1. View the first 5 rows of the DataFrame.
df.head()

# 2. Show where values are missing.
df.isnull()

# 3. Count missing values by column.
df.isnull().sum()

# 4. Drop rows with missing data.
df.dropna()

# 5. Re-create origingal 
df = pd.DataFrame(data)

# 6. Drop columns with missing data.
df.dropna(axis=1)

# 7. Re-create original
df = pd.DataFrame(data)

# 8. Fill missing Age values with the mean.
df.fillna(df['Age'].mean())

# 9. Fill missing City values with "Unknown".
df['City'] = df['City'].fillna('Unknown')
print(df)
```

Exercise Set 3 : Duplicates
Exercises 1-4

```python
import pandas as pd

data = {
"Name": ["Alice", "Bob", "Charlie", "Alice", "Bob", "Diana"],
"Age": [25, 30, 35, 25, 30, 40],
"City": ["NYC", "LA", "SF", "NYC", "LA", "NYC"],
}

df = pd.DataFrame(data)

# 1. Find which rows are duplicates (across all columns)
print("\nDuplicated rows (True if row is a duplicate of an earlier one):")
print(df.duplicated())

# # 2. Drop duplicate rows (keeping the first occurrence)
print("\nDataFrame with duplicates dropped (keep='first'):")
print(df.drop_duplicates(keep='first'))

# # 3. Drop duplicate rows (keeping only the last occurrence)
print("\nDataFrame with duplicates dropped (keep='last'):")
print(df.drop_duplicates(keep='last'))

# # 4. Drop ALL duplicates (remove both original and duplicate)
print("\nDataFrame with ALL duplicates removed (keep=False):")
print(df.drop_duplicates(keep=False))
```

Exercise Set 1 : Renaming Columns
Exercises 1-2

```python
import pandas as pd

data = {
"fname": ["Alice", "Bob", "Charlie"],
"lname": ["Smith", "Jones", "Brown"],
"ctry": ["USA", "Canada", "UK"],
}

df = pd.DataFrame(data)

# 1. Change fname to FirstName and lname to LastName.
df.rename(columns={
    'fname' : 'FirstName',
    'lname' : 'LastName'
}, inplace=True)
print(df)

# 2. Change all column names to: ['FirstName', 'LastName', 'Country']
df.columns = ['FirstName', 'LastName', 'Country']
print(df)
```

```python
import pandas as pd

data = {
"Name": ["Alice", "Bob", "Charlie"],
"Dept": ["HR", "Finance", "IT"],
"Sal": [60000, 70000, 80000],
}

df = pd.DataFrame(data)

# 1. Rename Dept to Department and Sal to Salary.
df.rename(columns={
    'Dept' : 'Department',
    'Sal' : 'Salary'
    }, inplace=True)
print(df)

# 2. Reorder columns so Salary appears first, followed by Name and Department.
df_reordered = df[['Salary', 'Name', 'Department']]
print(df_reordered)

# 3. Print the updated DataFrame.
print(df_reordered)
```

Exercise Set 3 : GroupBy and Agg
Exercises 1-4

```python
import pandas as pd

data = {
"Store": ["A", "A", "A", "B", "B", "C", "C", "C", "C"],
"Product": ["Apples", "Oranges", "Bananas", "Apples", "Bananas", "Oranges", "Apples", "Bananas", "Bananas"],
"Sales": [100, 80, 90, 120, 110, 60, 95, 85, 75],
"Profit": [30, 25, 20, 40, 35, 15, 32, 22, 18],
}

df = pd.DataFrame(data)
print(df.head())

# 1. Group the data by Store and use .agg() to find the average Sales and average Profit for each store
print(df.groupby('Store').agg({
    'Sales' : 'mean', 
    'Profit' : 'mean'}))

# 2. Group by Product and get the maximum Sales and minimum Profit
print(df.groupby('Product').agg({
    'Sales' : 'max',
     'Profit' : 'min' }))

# 3. Group by Store and return both the average Sales and maximum Profit, but name the columns 'Avg_Sales' and 'Max_Profit'
print(df.groupby('Store').agg(
    Avg_Sales = ('Sales', 'mean'),
    Max_Profit = ('Profit', 'max')
))

# 4. BONUS: Group by both Store and Product, and find the total Sales using .agg()
print(df.groupby(['Store', 'Product']).agg(total_Sales = ('Sales', 'sum')))
```

```python
import pandas as pd

df_customers = pd.DataFrame({
"CustomerName": ["Alice", "Bob", "Charlie"]
}, index=[101, 102, 103])

df_orders = pd.DataFrame({
"OrderTotal": [250, 300, 400]
}, index=[101, 102, 103])


# 1. Join df_orders to df_customers on the index.
print(df_customers.join(df_orders))


# 2. Add a new column to df_customers with city data from df_cities.
df_cities = pd.DataFrame({
    'City': ['NYC', 'LA', 'Chicago']
}, index=[101, 102, 103])
print(df_customers.join(df_cities))
```
